```python

# Drop the "Date" and "Time" columns
df = df.drop(["Date", "Time"], axis=1)

# Display the modified DataFrame
print(df)
```

```python
df
```

```python

# List of columns to drop
columns_to_drop = ["CO2_(dinning-room)", "Relative_humidity_(dinning-room)", "Lighting_(dinning-room)",
                   "Meteo_Sun_dusk", "Meteo_Sun_light_in_east_facade", "Meteo_Sun_light_in_south_facade",
                   "Meteo_Sun_irradiance", "Day_of_the_week"]

# Drop the specified columns
df = df.drop(columns=columns_to_drop)

# Display the modified DataFrame
print(df)
```

```python
df
```

```python
# Assuming your DataFrame is named 'df'
x = df.drop(columns=['Indoor_temperature_room'])
y = df['Indoor_temperature_room']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lir = LinearRegression()
lir.fit(x_train_scaled,y_train)
```

```python
pred = lir.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(pred,y_test)
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
pred = rf.predict(x_test)
```

```python
pred
```

```python
r2_score(y_test,pred)
```

```python
lg = lgb.LGBMRegressor()
```

```python
lg.fit(x_train,y_train)
```

```python
pred = lg.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
xg = xgb.XGBRegressor()
```

```python
xg.fit(x_train,y_train)
```

```python
pred = xg.predict(x_test)
```

```python
r2_score(y_test,pred)
```
